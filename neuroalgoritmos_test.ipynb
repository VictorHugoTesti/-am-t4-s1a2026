```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
import pickle
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Impartando Digest base_ibama

```python
with open ('/content/drive/MyDrive/Aprendizagem de Maquina/ibama.pkl', 'rb') as f:
  X_treinamento, Y_treinamento, X_teste, Y_teste = pickle.load(f)
```

```python
X_treinamento.shape, Y_treinamento.shape
```

```python
X_teste.shape, Y_teste.shape
```

# Pipeline Árvore de Decisão | Treemap [Fase 1 - base_ibama]

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore_ibama = DecisionTreeClassifier(criterion='entropy', random_state=0)
arvore_ibama.fit(X_treinamento, Y_treinamento)
DecisionTreeClassifier(ccp_alpha=0.0, class_weight=None, criterion='entropy',
                       max_depth=None, max_features=None, max_leaf_nodes=None,
                       min_samples_leaf=1, min_samples_split=2, min_weight_fraction_leaf=0.0,
                       random_state=0, splitter='best')
```

```python
arvore_previsoes = arvore_ibama.predict(X_teste)
arvore_previsoes
```

```python
print("Acurácia:", accuracy_score(Y_teste, arvore_previsoes))
print("\nMatrix de Confusão:\n", confusion_matrix(Y_teste, arvore_previsoes))
print("\nRelatório Final:\n", classification_report(Y_teste, arvore_previsoes))
```

# Pipeline Random Forest | Digest Pickle [Fase 2 - base_ibama]

```python
from sklearn.ensemble import RandomForestClassifier
from warnings import warn
```

```python
random_ibama = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
random_ibama.fit(X_treinamento, Y_treinamento)
RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='entropy', max_depth='auto',
                       min_impurity_decrease=0.0, min_samples_split=None,
                       min_samples_leaf=1, min_weight_fraction_leaf=0.0,
                       n_estimators=10, n_jobs=None, oob_score=False,
                       random_state=0, verbose=0, warm_start=False)
```

```python
random_previsores = random_ibama.predict(X_teste)
random_previsores
```

```python
print("Acurácia:", accuracy_score(Y_teste, random_previsores))
print("\nMatrix de Confusão:\n", confusion_matrix(Y_teste, random_previsores))
print("\nRelatório Final:\n", classification_report(Y_teste, random_previsores))
```

# Pipeline SVM | Digest [Fase 3 - base_ibama]

```python
from sklearn.svm import SVC
```

```python
svm_ibama = SVC(kernel='rbf', random_state=1, C=2.0)
svm_ibama.fit(X_treinamento, Y_treinamento)
SVC(C=2.0, break_ties=False, cache_size=200, class_weight=None,
    coef0=0.0, decision_function_shape='ovr', degree=3,
    kernel='rbf', max_iter=1, probability=False, random_state=1,
    shrinking=True, tol=0.001, verbose=False)
```

```python
svm_previsores = svm_ibama.predict(X_teste)
svm_previsores
```

```python
print("Acurácia:", accuracy_score(Y_teste, svm_previsores))
print("\nMatrix de Confusão:\n", confusion_matrix(Y_teste, svm_previsores))
print("\nRelatório Final:\n", classification_report(Y_teste, svm_previsores))
```

# Pipeline Preparação de Dados (KNN) | Digest [Fase 5 - base_ibama]

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
parametros = {'n_neighbors': [3, 5, 10, 20],
              'p': [1,2]}
```

```python
grid_ibama = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parametros)
grid_ibama.fit(X_treinamento, Y_treinamento)
melhor_parametro_ibama = grid_ibama.best_params_
melhor_resultado_ibama = grid_ibama.best_score_
```

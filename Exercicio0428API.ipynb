```python
import requests
import pandas as pd
import time

API_KEY = 'DFFS0TVTN2FMLCL8'

EMPRESAS = ['NFLX', 'AAP', 'GOOGL', 'AMZN', 'MSFT']
lista_emp = []

for simbolo in EMPRESAS:
  print(f'Extraindo dados de {simbolo}')
  url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={simbolo}&interval=5min&apikey={API_KEY}'
  r = requests.get(url)
  base = r.json()

  if 'Global Quote' in base:
    base_alpha = pd.DataFrame.from_dict([base['Global Quote']])
    base_alpha['symbol'] = simbolo
    lista_emp.append(base_alpha)

  time.sleep(3)

base_alpha_empresas = pd.concat(lista_emp, ignore_index=True)
cols_numerics = ['02. open', '03. high', '04. low', '05. price', '06. volume', '08. previous close', '09. change', '10. change percent']

base_alpha_empresas
```

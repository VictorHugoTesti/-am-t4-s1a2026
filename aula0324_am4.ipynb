# Bibliotecas e Funções de Python


```python
import math

def equacao2grau(a, b, c):
  delta = b**2 - 4*a*c

  if(delta < 0):
    return "Numero sem Raiz", delta

  x1 = (-b + math.sqrt(delta))/(2*a)
  x2 = (-b - math.sqrt(delta))/(2*a)

  return x1, x2
```

## Testes de Funções Matematicas

```python
eq = equacao2grau(1, -3, 2)
print(eq)
```

```python
import random

for i in range(5):
  a = random.randint(-5, 1)
  while a==0:
    a = random.randint(-5, 1)
  b = random.randint(-10, 10)
  c = random.randint(-10, 10)

eq += equacao2grau(a, b ,c)

print(eq)

```

```python
for i in range(18):
  if(i % 2 == 0):
    if(eq[i] != "Numero sem Raiz"):
      print("x1: [", eq[i], "]")
  else:
    if(eq[i] != "Numero sem Raiz"):
      print("x2: [", eq[i], "]\n")

```

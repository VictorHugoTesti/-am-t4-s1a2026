```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## Carregar Digest de base_credit

```python
import pickle
```

```python
with open('/content/drive/MyDrive/Aprendizagem de Maquina/credit.pkt', 'rb') as f:
  X_credit_treinamento, Y_credit_treinamento, X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, Y_credit_treinamento.shape
```

```python
X_credit_teste.shape, Y_credit_teste.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_credit_data = GaussianNB()
naive_credit_data.fit(X_credit_treinamento, Y_credit_treinamento)
```

## Criando Base Previsoria

```python
credit_predict = naive_credit_data.predict(X_credit_teste)
```

```python
credit_predict
```

```python
Y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(Y_credit_teste, credit_predict)
```

```python
confusion_matrix(Y_credit_teste, credit_predict)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm_credit = ConfusionMatrix(naive_credit_data)
cm_credit.fit(X_credit_treinamento, Y_credit_treinamento)
cm_credit.score(X_credit_teste, Y_credit_teste)
```

# Pipeline Árvore de Decisão | Treemap Base [Fase 2 credit_data]

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore_credit_data = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
arvore_credit_data.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
DecisionTreeClassifier(ccp_alpha=0.0, class_weight=None, criterion='entropy',
                       max_depth=None, max_features=None, max_leaf_nodes=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0,
                       random_state=0, splitter='best')
```

```python
previsoes_arvore = arvore_credit_data.predict(X_credit_teste)
```

```python
previsoes_arvore
```

```python
Y_credit_teste
```

```python
accuracy_score(Y_credit_teste, previsoes_arvore)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
print(classification_report(Y_credit_teste, previsoes_arvore))
```

```python
from sklearn import tree
previsoes_arvore = ['Income', 'Age', 'Loan']
fig = plt.figure(figsize=(25,20))
tree.plot_tree(arvore_credit_data, feature_names=previsoes_arvore, filled=True)
plt.show()
```

```python
cm_tree_credit = ConfusionMatrix(arvore_credit_data)
cm_tree_credit.fit(X_credit_treinamento, Y_credit_treinamento)
cm_tree_credit.score(X_credit_teste, Y_credit_teste)
```
